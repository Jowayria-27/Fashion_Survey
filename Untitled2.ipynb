```python
columns_to_fix = [f'FQ{i}' for i in list(range(1, 4)) + list(range(6, 11)) + [12]]

# Standardize responses (fix typos + handle case/whitespace)
response_mapping = {
    r'disgree': 'disagree',               # Fix 'disgree' typo
    r'strongly\s*disgree': 'strongly disagree',  # Fix 'strongly disgree' typo
    r'strongly\s*disagree': 'strongly disagree', # Ensure consistency
    r'disagree': 'disagree'               # Standardize
}

# Apply to all specified columns
for col in columns_to_fix:
    if col in df_gamified.columns:
        df_gamified[col] = (
            df_gamified[col]
            .astype(str)
            .str.strip()
            .str.lower()
            .replace(response_mapping, regex=True)
            .str.title()  # Capitalize first letters (optional)
        )
```

```python
df_game_demo=df_gamified_k.copy()
df_trad_demo=df_trad_k.copy()
df_game_demo.drop(['Survey Type','FQ1','FQ2', 'FQ3', 'FQ4', 'FQ6', 'FQ7','FQ8', 'FQ9','FQ10'],axis=1,inplace=True)
df_trad_demo.drop(['Survey Type','FQ1','FQ2', 'FQ3', 'FQ4', 'FQ6', 'FQ7','FQ8', 'FQ9','FQ10'],axis=1,inplace=True)
df_game_demo.head()
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
df_game_demo['Duration_sec'] = pd.to_timedelta(df_game_demo['Duration']).dt.total_seconds()

# Define feature types
numerical_features = ['PU', 'RF', 'FA', 'AE', 'Duration_sec']
categorical_features = ['Age', 'Gender', 'Education level']  # All treated as categorical

# Create preprocessing pipeline with one-hot encoding for all categorical variables
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(drop='first'), categorical_features)  # drop='first' to avoid dummy variable trap
    ])

# Apply preprocessing
processed_features = preprocessor.fit_transform(df_game_demo)


# Get feature names after one-hot encoding
encoder = preprocessor.named_transformers_['cat']
cat_feature_names = encoder.get_feature_names_out(categorical_features)
all_feature_names = numerical_features + list(cat_feature_names)
processed_df = pd.DataFrame(processed_features, columns=all_feature_names)
```

```python
processed_df.head()
```

```python
# Elbow method
sse = []
for k in range(1, min(6, len(processed_df))):  # Don't exceed number of samples
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(processed_features)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, len(sse)+1), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
k = 2
kmeans = KMeans(n_clusters=k, random_state=42)
processed_df['Cluster'] = kmeans.fit_predict(processed_features)

# Add cluster labels to original data
processed_df['Cluster'] = processed_df['Cluster'].astype('category')
```

```python
for col in processed_df.select_dtypes('number').columns:
    plt.figure()
    sns.boxplot(x='Cluster', y=col, data=processed_df)
```
