There is a typo with disagree and strongly disagree. Fix it

```python
columns_to_fix = [f'FQ{i}' for i in list(range(1, 4)) + list(range(6, 11)) + [12]]

# Standardize responses (fix typos + handle case/whitespace)
response_mapping = {
    r'disgree': 'disagree',               # Fix 'disgree' typo
    r'strongly\s*disgree': 'strongly disagree',  # Fix 'strongly disgree' typo
    r'strongly\s*disagree': 'strongly disagree', # Ensure consistency
    r'disagree': 'disagree'               # Standardize
}

# Apply to all specified columns
for col in columns_to_fix:
    if col in df_gamified.columns:
        df_gamified[col] = (
            df_gamified[col]
            .astype(str)
            .str.strip()
            .str.lower()
            .replace(response_mapping, regex=True)
            .str.title()  # Capitalize first letters (optional)
        )
df_gamified['Age'].replace('45976','11-15',inplace=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Map clusters to descriptive labels
combined_df['cluster'] = combined_df['cluster'].map({0: 'Strong Engagement', 1: 'Moderate Engagement'})

# Create a contingency table
cluster_dist = pd.crosstab(combined_df['Survey_Type'], combined_df['cluster'])

# Chi-square test
chi2, p, _, _ = chi2_contingency(cluster_dist)
print(f"Chi2: {chi2}, p-value: {p}")

# Plot stacked bar chart
ax = cluster_dist.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Cluster Distribution by Survey Type')
plt.xlabel('Survey Type')
plt.ylabel('Number of Participants')
plt.legend(title='Cluster')

# Add raw count annotations
for i, (index, row) in enumerate(cluster_dist.iterrows()):
    cumulative = 0
    for j, value in enumerate(row):
        cumulative += value
        ax.text(i, cumulative - value/2, str(value), ha='center', va='center', color='white', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors

def tune_dbscan(X, eps_range, min_samples_range, plot_k_dist=True):
    """
    Tune DBSCAN parameters using Silhouette Score and k-distance plot.

    Parameters:
    - X: numpy array or DataFrame of your data
    - eps_range: list or array of eps values to try
    - min_samples_range: list or array of min_samples values to try
    - plot_k_dist: whether to show the k-distance plot (True/False)

    Returns:
    - best_params: dict with 'eps', 'min_samples', 'score'
    """

    # Plot k-distance graph
    if plot_k_dist:
        min_samples_for_plot = max(min_samples_range)
        neighbors = NearestNeighbors(n_neighbors=min_samples_for_plot)
        neighbors_fit = neighbors.fit(X)
        distances, _ = neighbors_fit.kneighbors(X)
        k_distances = np.sort(distances[:, min_samples_for_plot - 1])
        plt.figure(figsize=(8, 4))
        plt.plot(k_distances)
        plt.xlabel('Points sorted by distance')
        plt.ylabel(f'{min_samples_for_plot}-NN distance')
        plt.title('k-distance graph')
        plt.grid()
        plt.show()

    # Grid search over eps and min_samples
    best_score = -1
    best_params = {'eps': None, 'min_samples': None, 'score': -1}

    for eps in eps_range:
        for min_samples in min_samples_range:
            try:
                db = DBSCAN(eps=eps, min_samples=min_samples)
                labels = db.fit_predict(X)
                # Silhouette score is only defined if there's more than 1 cluster
                if len(set(labels)) > 1 and len(set(labels)) < len(X):
                    score = silhouette_score(X, labels)
                    if score > best_score:
                        best_score = score
                        best_params = {
                            'eps': eps,
                            'min_samples': min_samples,
                            'score': score
                        }
            except Exception as e:
                print(f"Skipped eps={eps}, min_samples={min_samples}: {e}")

    print("\nBest parameters found:")
    print(f"eps = {best_params['eps']}, min_samples = {best_params['min_samples']}, silhouette score = {best_params['score']:.3f}")
    return best_params
# Replace this with your desired search ranges
eps_range = np.arange(0.1, 2.0, 0.1)
min_samples_range = range(3, 12)

# Run the tuning function
best_params = tune_dbscan(X.values, eps_range, min_samples_range)
```
