There is a typo with disagree and strongly disagree. Fix it

```python
columns_to_fix = [f'FQ{i}' for i in list(range(1, 4)) + list(range(6, 11)) + [12]]

# Standardize responses (fix typos + handle case/whitespace)
response_mapping = {
    r'disgree': 'disagree',               # Fix 'disgree' typo
    r'strongly\s*disgree': 'strongly disagree',  # Fix 'strongly disgree' typo
    r'strongly\s*disagree': 'strongly disagree', # Ensure consistency
    r'disagree': 'disagree'               # Standardize
}

# Apply to all specified columns
for col in columns_to_fix:
    if col in df_gamified.columns:
        df_gamified[col] = (
            df_gamified[col]
            .astype(str)
            .str.strip()
            .str.lower()
            .replace(response_mapping, regex=True)
            .str.title()  # Capitalize first letters (optional)
        )
```

```python
# prompt: I want FQ1 to FQ3 and from FQ6 to FQ10 to have values of -2,-1,0,1,2 for both df_gamified_k and df_trad_k

# Define the mapping for FQ1 to FQ3 and FQ6 to FQ10
response_mapping = {
    'Strongly Disagree': -2,
    'Disagree': -1,
    'Neutral': 0,
    'Agree': 1,
    'Strongly Agree': 2
}

# Apply the mapping to the specified columns for both dataframes
for col in [f'FQ{i}' for i in range(1, 4)] + [f'FQ{i}' for i in range(6, 11)]:
    if col in df_gamified_k.columns:
        df_gamified_k[col] = df_gamified_k[col].map(response_mapping)
    if col in df_trad_k.columns:
        df_trad_k[col] = df_trad_k[col].map(response_mapping)
```

Now, let's create a df with new columns that has the average of each point.

```python
PU_questions = ['FQ1', 'FQ2', 'FQ3']
RF_questions = ['FQ4', 'FQ6', 'FQ7']
FA_questions = ['FQ8', 'FQ9']
AE_questions = ['FQ10']

# Calculate average score per dimension
df_gamified_k['PU'] = df_gamified_k[PU_questions].mean(axis=1)
df_gamified_k['RF'] = df_gamified_k[RF_questions].mean(axis=1)
df_gamified_k['FA'] = df_gamified_k[FA_questions].mean(axis=1)
df_gamified_k['AE'] = df_gamified_k[AE_questions].mean(axis=1)
df_gamified_k['Overall_Engagement'] = df_gamified_k[['PU', 'RF', 'FA', 'AE']].mean(axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
df_game_test['Duration_sec'] = pd.to_timedelta(df_game_test['Duration']).dt.total_seconds()
df_game_test=df_game_test.drop('Duration', axis=1)
df_game_test.head()
df_game_test["Engagement"]=df_gamified_k["Overall_Engagement"]
df_game_test.head()

# Select features for clustering
features = df_game_test[['PU', 'RF', 'FA', 'AE', 'Duration_sec']]

# Standardize the features
scaled_data = StandardScaler().fit_transform(df_game_test)
```

```python
scaled_data
```

```python
from sklearn.cluster import KMeans

# Use the Elbow method to find a good number of clusters using WCSS.
# We'll set the upper limit of k to 20 this time,
# to factor in the possibility of having 6 to 10 customer groups.
# Setting a higher number would cause time to run to increase,
# so some optimizations needs to considered for very large datasets

# !!Important Note: scale the upper limit with the number of groups you are expecting


wcss = []
for k in range( 1, 4 ):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit_predict(scaled_data)
    wcss.append( kmeans.inertia_ )

plt.plot(range(1, 4), wcss, 'o-r', label="WCSS")
plt.title("Computing WCSS for KMeans++")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans.fit_predict(scaled_data)

# As it's difficult to visualise clusters when the data is high-dimensional - we'll use
# boxplots to help us see how the clusters are grouping the samples

features["cluster"] = cluster_labels
```

```python
for col in features.select_dtypes('number').columns:
    plt.figure()
    sns.boxplot(x='cluster', y=col, data=features)
```
